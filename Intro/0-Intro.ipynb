```python
import numpy as np
import matplotlib.pylab as plt
```

```python
x1=np.append(np.random.normal(1,.2,1000),np.random.normal(3,.5,1000))
x2=np.append(np.random.normal(1,.2,1000),np.random.normal(4,.5,1000))
X=np.array([x1,x2]).T
y=np.append(np.zeros(1000),np.ones(1000))
```

```python
X
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
```

```python
plt.scatter(X[:,0],X[:,1])
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test,y_train,y_test = train_test_split(X,y,test_size=.2,random_state=2)
```

```python
X
```

```python
y
```

```python
plt.scatter(X_train[:,0],X_train[:,1],c=y_train)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
from sklearn import svm
```

```python

clf = svm.SVC()  #Define a classifier
clf.fit(X_train,y_train)
y_pred=clf.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
```

```python
conf_mat=(confusion_matrix(y_test,y_pred))
```

```python
ConfusionMatrixDisplay(confusion_matrix=conf_mat,display_labels=True)
```

```python
pl=plt.imshow(conf_mat)
plt.colorbar(pl)
```

```python
x1=np.append(np.random.normal(1,1,1000),np.random.normal(3,1,1000))
x2=np.append(np.random.normal(1,1,1000),np.random.normal(2,1,1000))
X=np.array([x1,x2]).T
y=np.append(np.zeros(1000),np.ones(1000))

plt.scatter(X[:,0],X[:,1],c=y)
```

```python
X_train, X_test,y_train,y_test = train_test_split(X,y,test_size=.2,random_state=2)
```

```python
plt.scatter(X_train[:,0],X_train[:,1],c=y_train)
```

```python

clf = svm.SVC()  #Define a classifier

clf.fit(X_train,y_train)
y_pred=clf.predict(X_test)
```

```python
conf_mat=(confusion_matrix(y_test,y_pred))
print(conf_mat)
```

```python
pl=plt.imshow(conf_mat)
plt.colorbar(pl)
```
